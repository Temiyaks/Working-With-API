I"II be scaping the allbirds.com to get more data. However, the website  have 

provided an API for the products. 

In order to work with APIs in Python, we need tools that will make those request. In Python, the most common library for making requests and working 
with APIs is the requests library.

```python

import requests
```

```python
#the url of the API
url ="https://www.allbirds.com/products.json"
```

```python
print(url)
```

# **Observation**
Inspecting the url on the web , this shows is  a JSON object! As you can see, it is very similar to a python dictionary and is made up of key-value pairs. In fact, in order for us to parse through this and extract what we want from it, we will eventually turn it into a python dictionary object. Upon inspection, we can see that it looks like a nested dictionary. The outer dictionary has the keys ‘Products’ (with a value of a list that is made up of dictionaries, with each dictionary coresponding to a specific products information). 

```python
#GET request to retrieve data
response = requests.get(url)
response
```

```python
requests.get(url).text
```

From the code above, This shows what the response is to our HTTP request from the API. As you can see, it is a long python string that is in JSON format





The get() function returns a response object. We can use the response status_code attribute to receive the status code for our request to know if if it is succesfull or not.

```python
print(response.status_code)
```

From the code above, We received a ‘200’ code which tells us our request was successful

200 API status code  means  everything is okay.


### **Creating a Python Dictionary**


Since the response is in JSON format, we need to  convert it into a python dictionary. Requests library  has a built-in JSON decoder to convert our JSON object to dictionary. 

```python
#To convert to dict and  see the data we received back from the API
response_info = (response.json())
response_info
```

```python
type(response_info)
```

From the code above, the conversion has been made to dictionary.

Now that our response is in the form of a python dictionary, we can use what we know about python dictionaries to parse it and extract the information we need!

```python
response_info['products'][0]
```

## **Observation**


*  Product is the outside dictionary
*   Variant is a sub dictionary inside the products, and it contains list of dictionaries.




```python
data=[] # To hold the information
for response in response_info["products"]: #loop through the response info
  for item in response["variants"]: #loop through the sub dictionary inside products

#get the indormation and append
#get is a dictionary method to get sub dictionaries.
    data.append({
    "Product_ID": response.get('id'),
    "Title": response.get('title'),
    "Published_Date": response.get('published_at'),
    "Product_Type": response.get('product_type'),

    "ID": item.get('id'),
    "SKU": item.get('sku'),
    "Price": item.get('price'),
    "Available": item.get('available'),
    "Created": item.get('created_at'),
    "Updated": item.get('updated_at'),
    "Compare_At_Price": item.get('compare_at_price'),
    
    })
    
   
  
    
       
```

```python
data[0:2] #To see few of the information
```

Now we will simply transform our list(data) into a pandas data frame with the pd.Dataframe method.

```python
#to import pandas library
import pandas as pd
```

```python
#to convert to dataframe
product = pd.DataFrame(data)
```

```python
#re_aaranging the column
product = product[["ID","Title","Published_Date", "Product_ID","Product_Type", "SKU", "Price","Available", "Created", "Updated","Compare_At_Price"]]
```
